### 1. Features in a Jupyter Notebook 
Once your jupyter session is running these are some basic contents: 
- The `kernel` is the sub-process where your browser/jupyter session will send the commands, the commands will compute, and the output will be retrieved and displayed (or files saved/modified or whatever the operation was) 
- The notebook contains `cells` each cell is your place to add inputs (either code to run, or text, or comments)
    - You create cells by adding them with `a` or `b` or with the `Insert>Cell Above/Below` menu buttons 
    - You copy, cut, and paste with `c` `x` and `v` 
    - You change the cell type with the `Cell>Cell Type` options or with the drop-down that reads Code/Markdown/Raw 
- One advantage of notebooks is that they allow you to show code and output and report/comment in-line being great to developing or experimenting. On the other hand, if you have a stable version of your code it is preferrable to create a stand-alone script and run it directly on python (without the jupyter environment) 
- The notebook is not linear but anything you run stays in the memory of unflushed processes/variables/definitions, so every once in a while you may need to restart the kernel (button in the top menu bar) 
- A package or library is a series of sourced scripts/executables that are appended to your environment for you to use in your program, you need to import them for them to be available in your current script/execution, and if they are not installed you need to install them (don't worry likely all packages we need are already installed and if not it's something we only need to do once) 
- Typically all package `import` calls are done at the beginning 
- Next session will show you how to run code

### 3. Run the following code

```python
## This section imports the required packages
import numpy 
from phi.flow import * 
from tqdm.notebook import trange
```

```python
# This is a basic command printing output to screen
print("Hello!") 
```

```python
## This section defines a function that we will use later
##     the function name is mandelbrot and that's how we will call/run it later
##          it takes 2 input variables re and im - which are numbers 
##          the function calls other functions but we will ignore it for now
##     note, we are not executing the function, just naming it and defining what it does so that we can use it later 
def mandelbrot(re, im): 
    ### output = repeatedly iterate passing output as input in next call 
    ####          z,i are iteratively defined 
    ####          C = re+im (imaginary/complex number) defined at start
    ####          zn = z**2 + C
    ####           iterate(function: (initial value, stability condition), iterations, initial z, initial i, range function)
    output = iterate(lambda z, i: (z**2+ re + im*1j, i+(abs(z) < 2)), 100, 0, 0, range=trange)[1]
    return output
```

> Side topic: the function above is running a Mathematical equation called the Mandelbrot set 
> The function is in complex plane (imaginary numbers) 
> If the next iteration is stable (i.e. does not diverge to infinity) we will color it by yellow 
> If it diverges, a color is chosen depending how quickly it does so 
> This will be done for every 
$Z_{n+1} = Z_{n}^{2} + c$

Which is iteratively run step by step 

```python
## CenteredGrid runs our function in the space for every value in the coordinates
##            re and im here are the resolution (or number of points in which it will be divided,
##            within the bounds 
## Each of the points in the space will run our mandelbrot function in its full iterations (as defined above)
## then the collection of resulting points plot just shows the generated collection of results in the grid 


plot(CenteredGrid(mandelbrot, re=1000, im=1000, 
                  bounds= Box(re=(-2, 1), im=(-1.5,1.5))))
```

- Don't worry about understanding every call in the functions above, just know that you can modify parameters and run it again 
- But every time you run it, thousands of processes are running to compute every point to the sepcified precision
